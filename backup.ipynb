```python
def batch_export(collection, loops):
    count = db[collection].count()
    batch_size = count / loops
    batch_size = math.ceil(batch_size)

    for i in range(loops):
        skip = batch_size*i
        !mongoexport --uri="mongodb://127.0.0.1:27017/tweets"  --collection={collection} --skip={skip} --limit={batch_size} --out=D:\mongo\collections\{collection}_batch{i}.json
```
